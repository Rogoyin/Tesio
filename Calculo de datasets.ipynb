```python
for Base in Bases:
    df = pd.read_excel(Paths[PC] + Base + '.xlsx')           

    # Indice de positividad.
    Inversion = {1: 5, 2: 4, 3: 3, 4: 2, 5: 1}               
    df['ece_item_5'] = df['ece_item_5'].map(Inversion)        
    df['indice_positividad'] = df[Items_Positividad].sum(axis=1)

    # Índices.
    df['indice_progresismo'] = df[Items_SA_Progresistas].mean(axis=1)
    df['indice_progresismo_tiempo'] = df[Items_SA_Progresistas_Tiempo].mean(axis=1)
    df['indice_conservadurismo'] = df[Items_SA_Conservadores].mean(axis=1)
    df['indice_conservadurismo_tiempo'] = df[Items_SA_Conservadores_Tiempo].mean(axis=1)

    # Cambios de opinión
    df['cambio_op_con_der'] = (df[Items_CA_Conservadores_Derecha].subtract(df[Items_SA_Conservadores].values)).mean(axis=1)
    df['cambio_op_pro_der'] = (df[Items_CA_Progresistas_Derecha].subtract(df[Items_SA_Progresistas].values)).mean(axis=1)

    df['cambio_op_con_der_tiempo'] = (df[Items_CA_Conservadores_Derecha_Tiempo].subtract(df[Items_SA_Conservadores_Tiempo].values)).mean(axis=1)
    df['cambio_op_pro_der_tiempo'] = (df[Items_CA_Progresistas_Derecha_Tiempo].subtract(df[Items_SA_Progresistas_Tiempo].values)).mean(axis=1)               

    df['cambio_op_con_izq'] = (df[Items_CA_Conservadores_Izquierda].subtract(df[Items_SA_Conservadores].values)).mean(axis=1)
    df['cambio_op_pro_izq'] = (df[Items_CA_Progresistas_Izquierda].subtract(df[Items_SA_Progresistas].values)).mean(axis=1)

    df['cambio_op_con_izq_tiempo'] = (df[Items_CA_Conservadores_Izquierda_Tiempo].subtract(df[Items_SA_Conservadores_Tiempo].values)).mean(axis=1)
    df['cambio_op_pro_izq_tiempo'] = (df[Items_CA_Progresistas_Izquierda_Tiempo].subtract(df[Items_SA_Progresistas_Tiempo].values)).mean(axis=1)

    # Verificar que no hay valores nulos y proporcionar información específica
    for Column in Columnas_A_Crear:
        if df[Column].notnull().all(): 
            print(f"Hay valores nulos en la columna '{Column}' en las filas: {df.index[df[Column].isnull()].tolist()}")

    # Verificar que los valores estén dentro de un rango específico.
    if len(df[(df['indice_progresismo'] > 5) | (df['indice_progresismo'] < 1)]) == 0: 
        print("Ningún índice de progresismo puede dar más de 5 o menos de 1.")
    if len(df[(df['indice_conservadurismo'] > 5) | (df['indice_conservadurismo'] < 1)]) == 0:
        print("Ningún índice de conservadurismo puede dar más de 5 o menos de 1.")

    # Guardar df.
    df.to_excel(Paths[PC] + Base + '. Arreglado.xlsx', index=False)    
```
