```python
# Definir computadora en que se va a trabajar.
PC = 'PC_Depto'
```

```python
# Definir la base en que se va a operar.
Base = '3. Ballotage - Filtrado (sin outliers).xlsx'
```

```python
# Crear dataframe.
df = pd.read_excel(Paths[PC] + Base)
```

```python
# Llenar espacios en blanco o "NaN" con 0.
df = df.fillna(value = 0)
```

```python
# Dar formato a Twitter.
df['redes_uso'] = df['redes_uso'].apply(lambda x: x.replace('(x)', ''))
```

```python
# Dividir valores de "Redes sociales" en una lista.
df['redes_uso'] = df['redes_uso'].apply(lambda x: x.split(" - "))
```

```python
# Agregar columnas con redes.
Redes = ['twitter', 'facebook', 'instagram', 'threads', 'tiktok', 'youtube', 'telegram', 'whatsapp']

for Element in Redes:
    df[Element] = ' '
```

```python
# Agregar valores correspondientes a redes.
for Index, Row in df.iterrows():
    for Red in Redes:
        if Red in Row['redes_uso']:
            df.loc[Index, Red] = int(1)
        else:
            df.loc[Index, Red] = int(0)

# Transformar a int.
df[Redes] = df[Redes].astype(int)
```

```python
df[['redes_uso'] + Redes]
```

```python
# Reemplazar 0 con 'ninguno' para evitar errores.
df['medios_prensa'] = df['medios_prensa'].replace(0, 'ninguno')
```

```python
# Dividir valores de "Medios de prensa" en una lista.
df['medios_prensa'] = df['medios_prensa'].apply(lambda x: x.split(" - "))
```

```python
# Agregar columnas con medios.
Medios = ['ambito financiero', 'prensa obrera', 'diario universal', 'popular', 'izquierda diario', 'clarin', 'perfil', 'pagina 12', 'infobae', 'el cronista', 'la nacion', 'tiempo argentino', 'ninguno']

for Element in Medios:
    df[Element] = ' '
```

```python
# Agregar valores correspondientes a medios.
for Index, Row in df.iterrows():
    for Medio in Medios:
        if Medio in Row['medios_prensa']:
            df.loc[Index, Medio] = int(1)
        else:
            df.loc[Index, Medio] = int(0)

# Transformar a int.
df[Medios] = df[Medios].astype(int)
```

```python
# Seleccionamos solo las columnas de tipo string (objetos).
Columnas_Strings = ['genero', 'nacionalidad', 'provincia', 'e_social', 'niv_educativo',
                    'f_ingreso', 'inmueble_res', 'voto_2019', 'voto_PASO_2023', 'candidato_PASO_2023', 
                    'categoria_PASO_2023', 'votara_2023', 'afiliacion_pol']

# Creamos las variables dummies.
df_Dummies = df
for Columna in Columnas_Strings:
    df_Dummies = fr.Create_Dummy_Variables(df_Dummies, Columna)
```

```python
# Identificamos las columnas dummy generadas (todas las que no estaban en el DataFrame original).
Columnas_Dummies = df_Dummies.select_dtypes(include='bool').columns

# Convertimos las columnas dummy de tipo bool a int.
df_Dummies[Columnas_Dummies] = df_Dummies[Columnas_Dummies].astype(int)

df_Dummies
```

```python
# Combinar dos dataframes sin duplicados.
Columnas_Unicas_df_Dummies = [Columna for Columna in df_Dummies.columns if Columna not in df.columns]

# Combinamos df1 con solo las columnas únicas de df2.
df_Final = pd.concat([df, df_Dummies[Columnas_Unicas_df_Dummies]], axis=1)
```

```python
# Formato de variables.

# Eliminar acentos.
df_Final = fr.Replace_Values_In_Name_Columns(df_Final, '-', '_')

# Eliminar espacios.
df_Final = fr.Replace_Values_In_Name_Columns(df_Final, ' ', '_')
```
